```python
import numpy as np
import pandas as pd

from utils import get_feature_names

import matplotlib.pyplot as plt
import seaborn as sns
# Pretty plots
%matplotlib inline
sns.set_style('ticks')
plt.rcParams['figure.figsize'] = (10, 4)
plt.rcParams['axes.titlesize'] = 22
plt.rcParams['axes.labelsize'] = 20
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['legend.fontsize'] = 12
plt.rcParams['legend.title_fontsize'] = 12

import warnings
warnings.filterwarnings("ignore")
```

```python
feature_names = get_feature_names('concepts')
```

```python
scores = pd.concat([
    pd.DataFrame({
        'method': 'Dictionary',
        'metric': 'precision',
        'concept': feature_names,
        'score':[0.83, 0.80, 0.94, 0.33, 0.94, 0.13, 0.04, 0.00, 0.12]
    }),
    pd.DataFrame({
        'method': 'Dictionary',
        'metric': 'recall',
        'concept': feature_names,
        'score':[0.71, 0.98, 0.94, 0.33, 0.97, 0.72, 0.80, np.nan, 1.00]
    }),
    pd.DataFrame({
        'method': 'CRF',
        'metric': 'precision',
        'concept': feature_names,
        'score':[1.0, 0.93, 0.93, np.nan, 0.97, 0.5, 1.0, np.nan, 0.9]
    }),
    pd.DataFrame({
        'method': 'CRF',
        'metric': 'recall',
        'concept': feature_names,
        'score':[1.0, 0.86, 0.78, 0.0, 0.94, 0.33, 0.20, np.nan, 0.35]
    }),
    pd.DataFrame({
        'method': 'BERT',
        'metric': 'precision',
        'concept': feature_names,
        'score':[0.5, 0.98, 0.89, 0.5, 0.97, 0.65, 0.9, np.nan, 0.89]
    }),
    pd.DataFrame({
        'method': 'BERT',
        'metric': 'recall',
        'concept': feature_names,
        'score':[0.57, 0.9, 0.94, 0.17, 0.94, 0.51, 0.9, np.nan, 0.92]
    })
])
```

```python
sns.barplot(x='concept', 
            y='score', 
            hue='method', 
            data=scores[scores.metric=='precision'], 
            palette='Paired');

plt.title("Precision");
plt.ylabel("Score");
plt.xticks(rotation=45);
plt.legend(bbox_to_anchor=(1.0, 0.5));

plt.savefig('precision on test');
```

```python
sns.barplot(x='concept', 
            y='score', 
            hue='method', 
            data=scores[scores.metric=='recall'], 
            palette='Paired');

plt.title("Recall");
plt.ylabel("Score");
plt.xticks(rotation=45);
plt.legend(bbox_to_anchor=(1.0, 0.5));

plt.savefig('recall on test');
```
